```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display  # 用于在 Jupyter Notebook 显示 DataFrame

# List of ETFs
etf_list = ["SPY", "QQQ", "IVV", "VTI", "VOO", "DIA", "ARKK", "EFA", "IWM", "TLT"]

# Define the time period
start_date = "2024-01-01"
end_date = "2024-03-01"

# Fetch historical data for all ETFs at once
try:
    data = yf.download(etf_list, start=start_date, end=end_date, progress=False)
    volume_df = data["Volume"]
except Exception as e:
    print(f"Error retrieving data: {e}")
    volume_df = pd.DataFrame()  # 创建空 DataFrame 避免后续错误

# Drop rows where all values are NaN
volume_df.dropna(how="all", inplace=True)

# Plot daily volume changes for the top traded ETF
if not volume_df.empty:
    top_etf = volume_df.mean(numeric_only=True).idxmax()  # 找到交易量最高的 ETF
    
    plt.figure(figsize=(12, 6))
    plt.plot(volume_df.index, volume_df[top_etf], label=top_etf, marker='o', linestyle='-')
    plt.xlabel("Date")
    plt.ylabel("Trading Volume")
    plt.title(f"Daily Trading Volume of {top_etf}")
    plt.legend()
    plt.grid()
    plt.show()

# Display the dataframe with trading volumes
display(volume_df)  # Jupyter Notebook 显示表格
```
