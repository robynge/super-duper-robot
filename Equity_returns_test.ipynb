```python
# Install yfinance module (if not already installed)
!pip install yfinance

# Import necessary libraries
import yfinance as yf
import pandas as pd

# Download historical data for the S&P 500 index
sp500 = yf.Ticker("^GSPC")
sp500_data = sp500.history(period="5y")  # Retrieve data for the past 5 years

# Calculate daily returns
sp500_data['Daily Return'] = sp500_data['Close'].pct_change()

# Print the first few rows of the data
print(sp500_data.head())

# Visualize the daily returns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
plt.plot(sp500_data.index, sp500_data['Daily Return'], label='Daily Return')
plt.title('S&P 500 Daily Returns')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.legend()
plt.show()
```
